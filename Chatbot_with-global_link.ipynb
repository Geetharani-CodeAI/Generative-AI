```python
import google.generativeai as genai
import gradio

genai.configure(api_key = "API_Key")

model = genai.GenerativeModel("gemini-2.5-flash")

chat = model.start_chat(history = [{"role": "user", 
            "parts": "you are a beauty experts that specializes women grooming technique"}]
                       )
def CustomGemini(user_input):
    response = chat.send_message(user_input)
    return response.text

demo = gradio.Interface(fn =CustomGemini, inputs = "text", outputs = "text", title = "My Savings Assistant")
demo.launch(share=True)
```
