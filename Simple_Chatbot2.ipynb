```python
import google.generativeai as genai

genai.configure(api_key = "API_Key")

model = genai.GenerativeModel("gemini-2.5-flash")

chat = model.start_chat(history=[])  # history=[] means no previous messages

system_msg=input("What type of chatbot would you like to create")
chat.send_message(system_msg)

print("Your new assistant is ready! (type'quit()' to exit)\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "quit()":
        print("Goodbye!")
        break

    response = chat.send_message(user_input)

    print("\n Assistant:", response.text, "\n")
```
